```python
digits_data = pd.read_csv("../data/digits.csv")
digits_data.drop(['Unnamed: 0'], axis = 1, inplace = True)
digits_data
```

```python
fig = plt.figure(figsize=(6, 6))  # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(digits_data.iloc[i][:-1], cmap=plt.cm.binary, interpolation='nearest')
    # label the image with the target value
    ax.text(0, 7, str(digits_data.iloc[i][65]))
```

```python
fig, ax = plt.subplots()
digits_data['target'].value_counts().plot(ax=ax, kind='bar')
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Class Distribution")
```

```python
fig, ax = plt.subplots()
digits_data['target'].value_counts().plot.pie(ax=ax)
plt.title("Class Distribution")
```

```python
mean_distribution = digits_data.describe().loc['mean'][:-1]
std_distribution = digits_data.describe().loc['std'][:-1]
min_distribution = digits_data.describe().loc['min'][:-1]
max_distribution = digits_data.describe().loc['max'][:-1]
median_distribution = digits_data.describe().loc['50%'][:-1]
```

```python
features_range = range(1,65)
plt.bar(features_range, mean_distribution)
plt.xlabel("Features")
plt.ylabel("Mean distribution")
plt.title("Mean Distribution of Features")
```

```python
features_range = range(1,65)
plt.bar(features_range, std_distribution)
plt.xlabel("Features")
plt.ylabel("Std distribution")
plt.title("Std Distribution of Features")
```

```python
features_range = range(1,65)
plt.bar(features_range, median_distribution)
plt.xlabel("Features")
plt.ylabel("Median distribution")
plt.title("Median Distribution of Features")
```

```python
features_range = range(1,65)
plt.bar(features_range, min_distribution)
plt.xlabel("Features")
plt.ylabel("Min distribution")
plt.title("Min Distribution of Features")
```

```python
fig =plt.figure(figsize = (10,6))
digits_data.iloc[:,:-1].skew(axis=0).plot(kind='bar')
plt.xlabel("Pixels")
plt.ylabel("Skewness")
plt.title("Skewness of features")
```

```python
fig =plt.figure(figsize = (10,6))
digits_data.iloc[:,:-1].kurtosis(axis=0).plot(kind='bar')
plt.xlabel("Pixels")
plt.ylabel("Kurtosis")
plt.title("Kurtosis of features")
```
