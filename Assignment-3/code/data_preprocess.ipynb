```python
import pandas as pd
from scipy.io import arff
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import seaborn as sns
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
data1 = pd.DataFrame(data= np.c_[digits['data'], digits['target']],
                     columns= digits['feature_names'] + ['target'])
```

```python
digits_data = pd.read_csv("../data/digits.csv")
digits_data.drop(['Unnamed: 0'], axis = 1, inplace = True)
digits_data
```

```python
img_mat = digits_data.iloc[0].to_numpy()
img_mat = img_mat[:-1].reshape(8, 8)
plt.imshow(img_mat, cmap=plt.get_cmap('gray'))
plt.savefig("../visualizations/eda/image_digit.png")
```

```python
fig, ax = plt.subplots()
digits_data['target'].value_counts().plot(ax=ax, kind='bar')
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Class Distribution")
plt.savefig("../visualizations/eda/barplot_class_distrb.png")
```

```python
fig, ax = plt.subplots()
digits_data['target'].value_counts().plot.pie(ax=ax)
plt.title("Class Distribution")
plt.savefig("../visualizations/eda/pichart_class_distrb.png")
```

```python
plt.figure(figsize = (25,6))
dataplot = sns.heatmap(digits_data.corr(), cmap="YlGnBu", annot=True)
plt.show()
plt.savefig("../visualizations/eda/corr_map.png")
```

```python
mean_distribution = digits_data.describe().loc['mean'][:-1]
std_distribution = digits_data.describe().loc['std'][:-1]
min_distribution = digits_data.describe().loc['min'][:-1]
max_distribution = digits_data.describe().loc['max'][:-1]
median_distribution = digits_data.describe().loc['50%'][:-1]
```

```python
features_range = range(1,65)
plt.bar(features_range, mean_distribution)
plt.xlabel("Features")
plt.ylabel("Mean distribution")
plt.title("Mean Distribution of Features")
plt.savefig("../visualizations/eda/mean_distrib.png")
```

```python
features_range = range(1,65)
plt.bar(features_range, std_distribution)
plt.xlabel("Features")
plt.ylabel("Std distribution")
plt.title("Std Distribution of Features")
plt.savefig("../visualizations/eda/std_distrib.png")
```

```python
features_range = range(1,65)
plt.bar(features_range, median_distribution)
plt.xlabel("Features")
plt.ylabel("Median distribution")
plt.title("Median Distribution of Features")
plt.savefig("../visualizations/eda/median_distrib.png")
```

```python
features_range = range(1,65)
plt.bar(features_range, min_distribution)
plt.xlabel("Features")
plt.ylabel("Min distribution")
plt.title("Min Distribution of Features")
plt.savefig("../visualizations/eda/min_distrib.png")
```

```python
features_range = range(1,65)
plt.bar(features_range, max_distribution)
plt.xlabel("Features")
plt.ylabel("Max distribution")
plt.title("Max Distribution of Features")
plt.savefig("../visualizations/eda/max_distrib.png")
```

```python
fig =plt.figure(figsize = (10,6))
digits_data.iloc[:,:-1].skew(axis=0).plot(kind='bar')
plt.xlabel("Pixels")
plt.ylabel("Skewness")
plt.title("Skewness of features")
plt.savefig("../visualizations/eda/skewness.png")
```

```python
fig =plt.figure(figsize = (10,6))
digits_data.iloc[:,:-1].kurtosis(axis=0).plot(kind='bar')
plt.xlabel("Pixels")
plt.ylabel("Kurtosis")
plt.title("Kurtosis of features")
plt.savefig("../visualizations/eda/kurtosis.png")
```
